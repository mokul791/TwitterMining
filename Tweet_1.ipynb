### Search: tweets using a keyword

```python
# Query for internet

query = 'internet'
language = 'en'

search_results = api.search(q=query, lang=language)
for tweet in search_results:
    print(tweet.user.screen_name, "Tweeted: ", tweet.text, "Created_at: ", tweet.created_at, "Location: ", tweet.user.location)
```

### Search with Hashtag

```python
search_words = '#service'
date_since = '2021-02-01' # 'yyyy-mon-day'

search_hashtag = tweepy.Cursor(api.search,
                      q = search_words,
                      lang = 'en',
                      since = date_since).items(10)

for tweet in search_hashtag:
    print(tweet.text)
    
```
